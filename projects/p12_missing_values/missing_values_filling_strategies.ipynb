```python
df1 = df1.dropna()
print(df1.isnull().sum()) #All missing values dropped
print("Shape: ", df.shape) #Shape:  (344, 7) 11 rows dropped
```

```python
df2 = df.copy()
```

```python
df2["bill_length_mm"] = df2["bill_length_mm"].fillna(df2.groupby("species")["bill_length_mm"].transform("mean"))
df2["bill_depth_mm"] = df2["bill_depth_mm"].fillna(df2.groupby("species")["bill_depth_mm"].transform("mean"))
df2["flipper_length_mm"] = df2["flipper_length_mm"].fillna(df2.groupby("species")["flipper_length_mm"].transform("mean"))
df2["body_mass_g"] = df2["body_mass_g"].fillna(df2.groupby("species")["body_mass_g"].transform("mean"))
#Global mean does not respect fr the difference of characteristics between different features. It will mix length, width and depth, distorting the real values.
#While specific mean maintains some originality in the dataset
```

```python
df2["sex"] = df2["sex"].fillna(df2["sex"].mode()[0])
# Filling with mode assumes missing sex values follow the same distribution as observed data, preserving class balance but risking bias if missingness is systematic.
# Leaving as "Unknown" preserves information about missingness but introduces an extra category that may distort categorical analysis if not meaningful.
```

```python
print(df2.isnull().sum()) #All filled
```

```python
df2.to_csv(r"penguins_final.csv", index = False)
```
