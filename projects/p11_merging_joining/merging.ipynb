```python
import pandas as pd
import numpy as np
```

```python
customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Ali", "Sara", "John", "Ayesha"],
    "city": ["Lahore", "Karachi", "Islamabad", "Multan"]
})
```

```python
orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104, 105],
    "customer_id": [1, 2, 2, 99, 3],
    "amount": [250, 400, 150, 999, 700]
})
```

```python
#Original shapes
print(f"Rows: {customers.shape[0]} & Columns: {customers.shape[1]}") #Rows: 4 & Columns: 3
print(f"Rows: {orders.shape[0]} & Columns: {orders.shape[1]}") #Rows: 5 & Columns: 3
```

```python
#Common merge
df_merge = pd.merge(customers, orders, on = "customer_id")
print(f"Rows: {df_merge.shape[0]} & Columns: {df_merge.shape[1]}") #Rows: 4 & Columns: 5
print("\n")
print(df_merge)
# Only matching customer_id values appear in the result and rows with customer_id present in only one table are excluded due to mismatch.
# For example, Ayesha is missing because their customer_id does not exist in both datasets, and the orders table contains inconsistent customer_id values.
```

```python
#inner merge
df_inner = pd.merge(customers, orders, on = "customer_id", how = "inner")
print(f"Rows: {df_inner.shape[0]} & Columns: {df_inner.shape[1]}") #Rows: 4 & Columns: 5
print("\n")
print(df_inner)
#The matching values survive and mismatch ones disappear, similar to simple merge
```

```python
#left merge
df_left = pd.merge(customers, orders, on = "customer_id", how = "left")
print(f"Rows: {df_left.shape[0]} & Columns: {df_left.shape[1]}") #Rows: 5 & Columns: 5
print("\n")
print(df_left)
#The data of first one survives while the second is rendered NaN like this one has Ayesha's row while the above ones didn't
```

```python
#right merge
df_right = pd.merge(customers, orders, on = "customer_id", how = "right")
print(f"Rows: {df_right.shape[0]} & Columns: {df_right.shape[1]}") #Rows: 5 & Columns: 5
print("\n")
print(df_right)
#Keeps the second data while rendering NaN the first one's mismatched rows. Ayesha is NaN here
```

```python
#outer merge
df_outer= pd.merge(customers, orders, on = "customer_id", how = "outer")
print(f"Rows: {df_outer.shape[0]} & Columns: {df_outer.shape[1]}") #Rows: 6 & Columns: 5
print("\n")
print(df_outer)
# pd.merge() defaults to how="inner", so only matching customer_id values from both
# DataFrames are included. This is why the result matches an explicit inner join
# and excludes any rows with non-matching customer_id values.
```
