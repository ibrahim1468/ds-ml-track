```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
print(df.head())
```

```python
columns = ["sex", "smoker", "day", "time"]
for i in columns:
    print(f"{i}: {df[i].unique()}")
```

```python
#Filter by total_bill > 30
bill30 = df[df["total_bill"] > 30]
print(f"Rows: {bill30.shape[0]} & Columns: {bill30.shape[1]}") #Rows: 32 & Columns: 7
print("\n")
print(bill30.head())
```

```python
#Filter by gender female
gf = df[df["sex"] == "Female"]
print(f"Rows: {gf.shape[0]} & Columns: {gf.shape[1]}") #Rows: 87 & Columns: 7
print("\n")
print(gf.head())
```

```python
#filter by condition
mtip = df[(df["sex"] == "Male") & (df["tip"] > 5)]
print(f"Rows: {mtip.shape[0]} & Columns: {df.shape[1]}") #Rows: 14 & Columns: 7
print("\n")
print(mtip.head())
```

```python
#sorting total bill descending
df_by_bill = df.sort_values(by = "total_bill", ascending = False)
print(df_by_bill.head())
```

```python
#Sorting by days and in it by tip
df_by_day_tip = df.sort_values(by = ["day", "tip"], ascending = [True, False])
print(df_by_day_tip.head())
```

```python
print(df.groupby("smoker", observed=True).agg({
    "tip": ["mean", "count"],
    "total_bill": ["mean", "sum"]
}))
```
