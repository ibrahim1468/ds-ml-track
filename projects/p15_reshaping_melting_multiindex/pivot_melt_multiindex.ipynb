```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset("flights")
```

```python
df_wide = df.pivot(index="year", columns="month", values="passengers")

print(df_wide.shape) #(12, 12)
print(df_wide.head())
```

```python
df_long = df_wide.reset_index().melt(
    id_vars="year",
    var_name="month",
    value_name="passengers"
)

month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
               "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

df_long["month"] = pd.Categorical(
    df_long["month"], 
    categories=month_order, 
    ordered=True
)

df_long = df_long.sort_values(["year", "month"]).reset_index(drop=True)

print(df_long.shape)
print(df_long.head(12))
```

```python
df_grouped = df.groupby(["year", "month"])["passengers"].sum()
print(df_grouped)
```
