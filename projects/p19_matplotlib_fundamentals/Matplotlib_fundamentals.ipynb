```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(x, y, c = "r", linewidth = 2)

ax.set_title("Sine Plot")
ax.set_xlabel("x")
ax.set_ylabel("Sine of x")

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

fig, ax = plt.subplots(figsize = (8,5))
ax.plot(x, y1, c = "r", linestyle = "-", label = "Sine")
ax.plot(x, y2, c = "g", linestyle = "--", label = "Cose")
ax.plot(x, y3, c = "b", linestyle = ":", label = "Sine x Cose")

ax.set_title("Sin & Cos")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize = (15, 4), sharey = True)
ax[0].plot(x, y1, c = "r", linestyle = "--")
ax[0].set_title("Sine")
ax[0].set_xlabel("x")
ax[0].set_ylabel("Sine x")

ax[1].plot(x, y2, c = "g", linestyle = "-")
ax[1].set_title("Cose")
ax[1].set_xlabel("x")
ax[1].set_ylabel("Cose x")

ax[2].plot(x, y3, c = "b", linestyle = ":")
ax[2].set_title("Sine & Cos")
ax[2].set_xlabel("x")
ax[2].set_ylabel("Sine x Cos x")

fig.suptitle("Comparing Three Functions")
plt.tight_layout()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 12, 38]
x = np.random.rand(50) * 10
y = x + np.random.randn(50) * 2

fig, ax = plt.subplots(1, 2, figsize = (12, 5))

ax[0].bar(categories, values)
ax[0].set_title("Bar Chart")
ax[0].set_xlabel("Categories")
ax[0].set_ylabel("Values")

ax[1].scatter(x, y)
ax[1].set_title("Scatter Chart")
ax[1].set_xlabel("X")
ax[1].set_ylabel("Y")

plt.tight_layout()
plt.show()
```

```python
x = np.arange(1, 11)
temperature = np.array([15,17,20,23,26,29,31,30,27,22])
rainfall = np.array([120,100,80,50,30,10,5,8,40,90])

fig, ax1 = plt.subplots(figsize = (9, 5))

ax1.plot(x, temperature, c = "r", label = "Temperature")
ax1.set_xlabel("Month")
ax1.set_ylabel("Temperature (°C)", c = "r")

ax2 = ax1.twinx()
ax2.plot(x, rainfall, c = "b", label = "Rainfall")
ax2.set_ylabel("Rainfall (mm)", c = "b")

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax1.legend(lines1 + lines2, labels1 + labels2)

plt.title("Temperature vs Rainfall")
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize = (8, 5))
ax.plot(x, y, c = "g")
idx = np.argmax(y)
ax.annotate("Maximum", xy = (x[idx], y[idx]), xytext = (x[idx] - 0.5, y[idx] - 1), arrowprops = dict(arrowstyle="->"))
plt.tight_layout()
plt.show()
```

```python
x = np.arange(1, 11)
y = 2 ** x

fig, ax = plt.subplots(1, 2, figsize = (12, 5))
ax[0].plot(x, y, c = "g")
ax[0].set_title("Without Log")
ax[0].set_xlabel("X")
ax[0].set_ylabel("Y")

ax[1].plot(x, y, c = "r")
ax[1].set_yscale("log")
ax[1].set_title("Log Scale")
ax[1].set_xlabel("X")
ax[1].set_ylabel("Log Y")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1 * y2

fig = plt.figure(figsize=(10, 6))
gs = fig.add_gridspec(2, 2)

ax1 = fig.add_subplot(gs[0, :])
ax1.plot(x, y1, color="royalblue")
ax1.set_title("sin(x)")

ax2 = fig.add_subplot(gs[1, 0])
ax2.plot(x, y2, color="crimson")
ax2.set_title("cos(x)")

ax3 = fig.add_subplot(gs[1, 1])
ax3.plot(x, y3, color="forestgreen")
ax3.set_title("sin(x) × cos(x)")

plt.tight_layout()
plt.show()
```
