```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
n = 80

df = pd.DataFrame({
    "category": np.random.choice(["A", "B", "C", "D"], size=n),
    "value": np.random.randn(n) * 10 + 50
})

df.head()
```

```python
sns.set_theme()
sns.histplot(data = df, x = "value")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.histplot(data = df, x = "value")
plt.show()
```

```python
np.random.seed(42)
n = 100

group_x = np.random.randn(n) * 10 + 50
group_y = np.random.randn(n) * 8 + 65

df = pd.DataFrame({
    "group": ["X"] * n + ["Y"] * n,
    "value": np.concatenate([group_x, group_y])
})

df.head()
```

```python
sns.set_style("whitegrid")
sns.kdeplot(data = df, x = "value", hue = "group")
sns.rugplot(data = df, x = 'value', hue = "group")
plt.plot()
```

```python
np.random.seed(42)
stores = (
    ["North"] * 120 +
    ["South"] * 80 +
    ["East"] * 50 +
    ["West"] * 30
)

daily_sales = np.concatenate([
    np.random.normal(loc=500, scale=50, size=120),
    np.random.normal(loc=700, scale=60, size=80),
    np.random.normal(loc=900, scale=70, size=50),
    np.random.normal(loc=300, scale=40, size=30)
])

df = pd.DataFrame({
    "store": stores,
    "daily_sales": daily_sales
})

df = df.sample(frac=1, random_state=42).reset_index(drop=True)
df.head()
```

```python
sns.countplot(data = df, x = "store")
plt.show()
```

```python
sns.barplot(data = df, x = "store", y = "daily_sales")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.boxplot(
    data=df,
    x="store",
    y="daily_sales",
    ax=axes[0, 0]
)
axes[0, 0].set_title("Boxplot")

sns.violinplot(
    data=df,
    x="store",
    y="daily_sales",
    ax=axes[0, 1]
)
axes[0, 1].set_title("Violin Plot")

sns.stripplot(
    data=df,
    x="store",
    y="daily_sales",
    jitter=True,
    alpha=0.7,
    ax=axes[1, 0]
)
axes[1, 0].set_title("Strip Plot")

sns.swarmplot(
    data=df,
    x="store",
    y="daily_sales",
    size=4,
    ax=axes[1, 1]
)
axes[1, 1].set_title("Swarm Plot")

plt.tight_layout()
plt.show()
```

```python
np.random.seed(42)

days = np.arange(1, 31)
records = []

for day in days:
    for sensor in ["A", "B"]:
        base = 20 + day * 0.5 if sensor == "A" else 25 + day * 0.4

        for _ in range(5):  # 5 readings per day per sensor
            reading = np.random.normal(loc=base, scale=2)
            records.append([day, sensor, reading])

df = pd.DataFrame(records, columns=["day", "sensor", "reading"])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.lineplot(
    data=df,
    x="day",
    y="reading",
    hue="sensor",
    ax=axes[0]
)
axes[0].set_title("Line Plot")

sns.scatterplot(
    data=df,
    x="day",
    y="reading",
    hue="sensor",
    ax=axes[1]
)
axes[1].set_title("Scatter Plot")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 300

x = np.random.normal(50, 10, n)
y = 0.8 * x + np.random.normal(0, 5, n)
z = -0.5 * x + np.random.normal(0, 4, n)

group = np.random.choice(["A", "B", "C"], size=n)

df = pd.DataFrame({
    "x": x,
    "y": y,
    "z": z,
    "group": group
})
```

```python
plt.figure(figsize=(6, 5))

sns.heatmap(
    df[["x", "y", "z"]].corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(
    df,
    hue="group",
    corner=True
)

plt.show()
```

```python
sns.jointplot(
    data=df,
    x="x",
    y="y",
    hue="group",
    height=7
)

plt.show()
```

```python
np.random.seed(42)

records = []

means = {
    ("East", "A"): 120,
    ("East", "B"): 180,
    ("East", "C"): 240,
    ("West", "A"): 150,
    ("West", "B"): 210,
    ("West", "C"): 170,
}

for region in ["East", "West"]:
    for product in ["A", "B", "C"]:
        for _ in range(40):
            revenue = np.random.normal(means[(region, product)], 15)
            quarter = np.random.choice(["Q1", "Q2"])  # Optional 3rd category
            records.append([region, product, revenue, quarter])

df = pd.DataFrame(records, columns=["region", "product", "revenue", "quarter"])
```

```python
sns.catplot(
    data=df,
    x="product",
    y="revenue",
    col="region",
    kind="box",
    height=5,
    aspect=1
)

plt.show()
```

```python
sns.catplot(
    data=df,
    x="product",
    y="revenue",
    row="quarter",
    col="region",
    kind="box",
    height=4,
    aspect=1
)

plt.show()
```

```python
np.random.seed(42)

n = 100

# Group A
x_a = np.random.uniform(0, 10, n)
y_a = 2 * x_a + np.random.normal(0, 2, n)

# Group B
x_b = np.random.uniform(0, 10, n)
y_b = 0.3 * x_b + np.random.normal(0, 2, n)

df = pd.DataFrame({
    "x": np.concatenate([x_a, x_b]),
    "y": np.concatenate([y_a, y_b]),
    "group": ["A"] * n + ["B"] * n
})
```

```python
sns.lmplot(
    data=df,
    x="x",
    y="y",
    hue="group",
    height=6,
    aspect=1.2
)
```

```python
sns.lmplot(
    data=df,
    x="x",
    y="y",
    col="group",
    height=6,
    aspect=1
)
```
